## Matplotlib example Workflow

```python
#Step 1 -> Import Matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt 

#Step 2 -> Prepare Datas
x= [1 , 2 , 3 , 4]
y= [11 , 22 , 33 , 44]

#Step 3 -> Setup Plot
fig , ax = plt.subplots(figsize=(10 , 5)) # figsize=(length , width)

#Step 4 -> Plot Datas
ax.plot(x , y)

#Step 5 -> Customize Plot
ax.set(title  = "Simple Plot" ,
       xlabel = "x-axis" ,
       ylabel = "y-axis")

#Step 6 -> Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

### Two options for subplots

```python
#Subplots (option 1)
fig , ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows=2,
                                                 ncols=2,
                                                 figsize=(10,5))
#Plot to each different axis
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10) , np.random.random(10))
ax3.bar(nut_butter_prices.keys() , nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

plt.show()
```

```python
#Subplots (option 2)
fig , ax= plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))
#Plot to each different index
ax[0 , 0].plot(x , x/2)
ax[0 , 1].scatter(np.random.random(10) , np.random.random(10))
ax[1 , 0].bar(nut_butter_prices.keys() , nut_butter_prices.values())
ax[1 , 1].hist(np.random.randn(1000))

plt.show()
```

## Which one you should use? (pyplot vs matplotlib OO method)
* When plotting something quickly , okay to use pyplot method
* When plotting more advanced , use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target')
plt.show()
```

```python
fig , ax = plt.subplots(figsize=(10 , 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)
plt.show()
ax.set_xlim([45 , 100])
```

```python
over_50.target.values
```

```python
# OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line for mean cholestrol
ax.axhline(over_50["chol"].mean(), linestyle='--', color='red', label='Mean Cholestrol')

# Optional: Add legend again if you want the line labeled
ax.legend()

# Show the plot
plt.show()
```

```python
#Subplot of age , chol , thalach
fig , (ax0 , ax1) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10 , 10),
                                 sharex = True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a meanline to ax0
ax0.axhline(y=over_50["chol"].mean(), linestyle='--')

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(), linestyle='--')

#Add a title to the figure
fig.suptitle("Heart Disease Analysis" , fontsize=16 , fontweight="bold")

plt.show()
```
