## Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y)
plt.show()
```

```python
#1st method
fig = plt.figure() #creates a figure
ax = fig.add_subplot() #add some axes
plt.show()
```

```python
#2nd method
fig = plt.figure() #creates a figure
ax = fig.add_axes([1,1,1,1]) #add some axes
ax.plot(x,y); #add some data
plt.show()
```

```python
#3rd method (recommended)
fig , ax = plt.subplots() #creates both the figure and axes in one line
ax.plot(x,[50 , 100 , 250 ,300]) #add some data
print(type(fig) , type(ax))
plt.show()
```

## Matplotlib example Workflow

```python
#Step 1 -> Import Matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt 

#Step 2 -> Prepare Datas
x= [1 , 2 , 3 , 4]
y= [11 , 22 , 33 , 44]

#Step 3 -> Setup Plot
fig , ax = plt.subplots(figsize=(10 , 5)) # figsize=(length , width)

#Step 4 -> Plot Datas
ax.plot(x , y)

#Step 5 -> Customize Plot
ax.set(title  = "Simple Plot" ,
       xlabel = "x-axis" ,
       ylabel = "y-axis")

#Step 6 -> Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

## Making Figures with NumPy arrays

We want:
* Line Plot
* Scatter Plot
* Bar Plot
* SubPlots
* Histogram

```python
import numpy as np
```

```python
#Create some Datas
x = np.linspace(0 , 10 , 100)
x[:10]
```

```python
#Plot the Data and create a line plot
fig , ax = plt.subplots()
ax.plot(x , x**2)
plt.show()
```

```python
# Use same data to create a Scatter Plot
fig , ax = plt.subplots()
ax.scatter(x , np.exp(x))
plt.show()
```

```python
#Another Scatter Plot
fig , ax = plt.subplots()
ax.scatter(x , np.sin(x))
plt.show()
```

```python
#Make a plot from a Dictionary
nut_butter_prices = {"Almond Butter" : 10 ,
                     "Peanut Butter" :  8 ,
                     "Cashew Butter" : 12}
fig , ax = plt.subplots()
ax.bar(nut_butter_prices.keys() , nut_butter_prices.values())

plt.show()
```

```python
fig , ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()) , list(nut_butter_prices.values()))

plt.show()
```

```python
#Make some data for histograms
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist(x)
plt.show()
```

### Two options for subplots

```python
#Subplots (option 1)
fig , ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows=2,
                                                 ncols=2,
                                                 figsize=(10,5))
#Plot to each different axis
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10) , np.random.random(10))
ax3.bar(nut_butter_prices.keys() , nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

plt.show()
```

```python
#Subplots (option 2)
fig , ax= plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(10,5))
#Plot to each different index
ax[0 , 0].plot(x , x/2)
ax[0 , 1].scatter(np.random.random(10) , np.random.random(10))
ax[1 , 0].bar(nut_butter_prices.keys() , nut_butter_prices.values())
ax[1 , 1].hist(np.random.randn(1000))

plt.show()
```
